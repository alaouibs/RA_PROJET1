```python
from OpenGL.GL import *
from OpenGL.GLU import *
import pylab as pl
from OpenGL.GLUT import *
from biosppy import storage
import numpy as np
import sys
import time

step = 0;
xmin = 0
xmax = 1000
ymin = -0.5
ymax = 5
# load raw ecg signal
signal, mdata = storage.load_txt('ecg.txt')
```

```python
# Some api in the chain is translating the keystrokes to this binary string
# so instead of saying: ESCAPE = 27, we use the following.
ESCAPE = b'\x1b'

# Number of the glut window.
window = 0

def init():
    # Commands # 1
    glEnable(GL_POINT_SMOOTH) # 1.1
    glEnable(GL_LINE_SMOOTH)  # 1.2
    glEnable(GL_BLEND)        # 1.3
    glBlendFunc(GL_SRC_ALPHA, GL_ONE_MINUS_SRC_ALPHA) # 1.4
    # Commands # 2
    glClearColor(1.0, 1.0, 1.0, 1.0) # 2.1
    gluOrtho2D(0, 1000.0, -0.5, 5.0) # 2.2
```

```python
def func_x2_sin(signal):
    vertices = [[x, (signal[x]-2047)*0.01] for x in range(len(signal)-1)]
    return np.array(vertices)
```

```python
def plot_func():
    
    
    global step
    global xmin
    global xmax
    global ymin
    global ymax
    global signal
    global mdata
    
    Fs = mdata['sampling_rate']
    N = len(signal)  # number of samples
    T = (N - 1) / Fs  # duration
    vertices = func_x2_sin(signal[(step % N):((1000 + step) % N)])

    # Commands # 3
    glClear(GL_COLOR_BUFFER_BIT) # 3.1
    glColor3f(0.0, 0.0, 0.0)     # 3.2
    glPointSize(3.0)             # 3.3
    glLineWidth(1.0)             # 3.4
    
        
    for i in range(25):
        glBegin(GL_LINES);        
        glColor3f(0.737255,0.560784,0.560784)
        glVertex2f(xmin, ymin + i*(ymax-ymin)/25)
        glVertex2f(xmax, ymin + i*(ymax-ymin)/25)        
        glVertex2f(i*(xmax-xmin)/25, ymin)
        glVertex2f(i*(xmax-xmin)/25, ymax)
        glEnd()
        if i%5 == 0:
            print(i*(ymax-ymin)/25)
            glBegin(GL_LINES);        
            glColor3f(0.737255,0.560784,0.560784)
            glVertex2f(xmin, ymin + i*(ymax-ymin)/25)
            glVertex2f(xmax, ymin + i*(ymax-ymin)/25)        
            glVertex2f(i*(xmax-xmin)/25, ymin)
            glVertex2f(i*(xmax-xmin)/25, ymax)
            glEnd()

            
        
        #declare Pink = color red 0.737255 green 0.560784 blue 0.560784
    # Commands # 5
    for i in range(len(vertices)-1):
        glBegin(GL_LINES)
        glColor3f(0.0, 0.0, 0.0)
        glVertex2fv(vertices[i])
        glVertex2fv(vertices[i+1])
        glEnd()
        
    # Commands # 6
    for i in range(len(vertices)):
        glBegin(GL_POINTS)
        glColor3f(0.0, 0.0, 0.0)
        glVertex2fv(vertices[i])
        glEnd()    
        # Commands # 7
    # since this is double buffered, swap the buffers to display what just got drawn.
    glutSwapBuffers()
    step = step + 12
```

```python
# The function called whenever a key is pressed. Note the use of Python tuples to pass in: (key, x, y)  
def keyPressed(*args):
    print(args[0])
    # If escape is pressed, kill everything.
    if args[0]==ESCAPE:
        glutDestroyWindow(window)
        sys.exit(0)
```

```python
def main():
    global window
    glutInit(())
    glutInitDisplayMode(GLUT_SINGLE|GLUT_RGB)
    glutInitWindowPosition(50,50)
    glutInitWindowSize(500,500)
    glutCreateWindow(b"Function Plotter")
    glutDisplayFunc(plot_func)
    # When we are doing nothing, redraw the scene.
    glutIdleFunc(plot_func)
    # Register the function called when the keyboard is pressed.  
    glutKeyboardFunc(keyPressed)
    # Initialization
    init()
    # Main drawing loop
    glutMainLoop()
```

```python
# Print message to console, and kick off the main to get it rolling.
print("Hit ESC key to quit.")
main()
```
